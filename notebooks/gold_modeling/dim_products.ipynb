```python
from pyspark.sql import functions as F

silver = "/Volumes/ecommerce_cat/ecommerce_schema/silver"
gold   = "/Volumes/ecommerce_cat/ecommerce_schema/gold"
```

```python
df = spark.read.format("delta").load(f"{silver}/products")
```

```python
dim_products = df.select(
    "product_id",
    "product_category_name",
    "product_weight_g",
    "product_length_cm",
    "product_height_cm",
    "product_width_cm"
).dropDuplicates(["product_id"])
```

```python
dim_products.write.format("delta").mode("overwrite").save(f"{gold}/dim_products")
```

```python
%sql
CREATE OR REPLACE VIEW ecommerce_cat.ecommerce_schema.dim_products AS
SELECT * FROM delta.`/Volumes/ecommerce_cat/ecommerce_schema/gold/dim_products`;
```
