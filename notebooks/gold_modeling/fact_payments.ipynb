```python
from pyspark.sql import functions as F

silver = "/Volumes/ecommerce_cat/ecommerce_schema/silver"
gold   = "/Volumes/ecommerce_cat/ecommerce_schema/gold"
```

```python
df = spark.read.format("delta").load(f"{silver}/order_payments")

fact_payments = df.select(
    "order_id",
    "payment_type",
    "payment_installments",
    "payment_value"
)
```

```python
fact_payments.write.format("delta").mode("overwrite").save(f"{gold}/fact_payments")
```

```python
%sql
CREATE OR REPLACE VIEW ecommerce_cat.ecommerce_schema.fact_payments AS
SELECT * FROM delta.`/Volumes/ecommerce_cat/ecommerce_schema/gold/fact_payments`;
```
