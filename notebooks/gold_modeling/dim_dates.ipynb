```python
from pyspark.sql import functions as F

silver = "/Volumes/ecommerce_cat/ecommerce_schema/silver"
gold   = "/Volumes/ecommerce_cat/ecommerce_schema/gold"
```

```python
from pyspark.sql.functions import explode, sequence, to_date, lit, col, year, month, dayofmonth

start_date = to_date(lit("2016-01-01"))
end_date   = to_date(lit("2018-12-31"))

df_dates = (
    spark.createDataFrame([(1,)])
    .select(
        explode(
            sequence(start_date, end_date, lit(1).cast("interval day"))
        ).alias("date")
    )
)

dim_dates = (
    df_dates
    .withColumn("year", year("date"))
    .withColumn("month", month("date"))
    .withColumn("day", dayofmonth("date"))
)
```

```python
dim_dates.write.format("delta").mode("overwrite").save(f"{gold}/dim_dates")
```

```python
%sql
CREATE OR REPLACE VIEW ecommerce_cat.ecommerce_schema.dim_dates AS
SELECT * FROM delta.`/Volumes/ecommerce_cat/ecommerce_schema/gold/dim_dates`;
```
