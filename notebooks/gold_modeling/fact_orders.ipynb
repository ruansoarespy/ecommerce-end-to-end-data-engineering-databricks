```python
from pyspark.sql import functions as F

silver = "/Volumes/ecommerce_cat/ecommerce_schema/silver"
gold   = "/Volumes/ecommerce_cat/ecommerce_schema/gold"
```

```python
df = spark.read.format("delta").load(f"{silver}/orders")
```

```python
fact_orders = df.select(
    "order_id",
    "customer_id",
    "order_status",
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date"
)
```

```python
fact_orders.write.format("delta").mode("overwrite").save(f"{gold}/fact_orders")
```

```python
%sql
CREATE OR REPLACE VIEW ecommerce_cat.ecommerce_schema.fact_orders AS
SELECT * FROM delta.`/Volumes/ecommerce_cat/ecommerce_schema/gold/fact_orders`;
```
