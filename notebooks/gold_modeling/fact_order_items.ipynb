```python
from pyspark.sql import functions as F

silver = "/Volumes/ecommerce_cat/ecommerce_schema/silver"
gold   = "/Volumes/ecommerce_cat/ecommerce_schema/gold"
```

```python
df = spark.read.format("delta").load(f"{silver}/order_items")

fact_order_items = df.select(
    "order_id",
    "order_item_id",
    "product_id",
    "seller_id",
    "price",
    "freight_value"
)
```

```python
fact_order_items.write.format("delta").mode("overwrite").save(f"{gold}/fact_order_items")
```

```python
%sql
CREATE OR REPLACE VIEW ecommerce_cat.ecommerce_schema.fact_order_items AS
SELECT * FROM delta.`/Volumes/ecommerce_cat/ecommerce_schema/gold/fact_order_items`;
```
