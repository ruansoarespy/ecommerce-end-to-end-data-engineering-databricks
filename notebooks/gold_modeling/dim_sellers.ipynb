```python
from pyspark.sql import functions as F

silver = "/Volumes/ecommerce_cat/ecommerce_schema/silver"
gold   = "/Volumes/ecommerce_cat/ecommerce_schema/gold"
```

```python
df = spark.read.format("delta").load(f"{silver}/sellers")

dim_sellers = df.select(
    "seller_id",
    "seller_city",
    "seller_state"
).dropDuplicates(["seller_id"])

dim_sellers.write.format("delta").mode("overwrite").save(f"{gold}/dim_sellers")
```

```python
%sql
CREATE OR REPLACE VIEW ecommerce_cat.ecommerce_schema.dim_sellers AS
SELECT * FROM delta.`/Volumes/ecommerce_cat/ecommerce_schema/gold/dim_sellers`;
```
