```python
import shutil
import os
import kagglehub

# Caminho da sua camada bronze
BRONZE_PATH = "/Volumes/ecommerce_cat/ecommerce_schema/bronze"

# Baixar o dataset
path = kagglehub.dataset_download("olistbr/brazilian-ecommerce")

# Se o path for um arquivo zip, você pode querer extrair
if path.endswith(".zip"):
    import zipfile
    with zipfile.ZipFile(path, 'r') as zip_ref:
        zip_ref.extractall(BRONZE_PATH)
else:
    # Caso seja apenas um arquivo, move para a pasta bronze
    shutil.move(path, os.path.join(BRONZE_PATH, os.path.basename(path)))

print("Arquivos disponíveis em:", BRONZE_PATH)
```
